```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
from langchain.llms.huggingface_pipeline import HuggingFacePipeline  # Updated import for HuggingFacePipeline
from langchain_core.prompts import PromptTemplate
model_name: str = "microsoft/phi-2"

# Check for GPU
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Load the model and tokenizer
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    low_cpu_mem_usage=True
).to(device)  # Move model to GPU (if available)

tokenizer = AutoTokenizer.from_pretrained(model_name)

# Define the model pipeline with device argument and truncation
model_pipeline = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    max_length=1024,
    pad_token_id=tokenizer.eos_token_id,
    truncation=True,  # Enable truncation
    device=0  # Use GPU (if available)
)

# Define generation arguments
gen_kwargs = {"temperature": 0.9}

# Create the HuggingFacePipeline
llm = HuggingFacePipeline(pipeline=model_pipeline, model_kwargs=gen_kwargs)

# Generate text
# output = llm.invoke('hello, how are you?')
# print(output)

prompt_template = PromptTemplate.from_template(
    """Instruct: {prompt}\nOutput:"
"""
)
user_prompt = "Write a detaitled analogy between mathematics and a lighthouse"
messages = prompt_template.format_prompt(prompt=user_prompt).to_messages()
output = llm.invoke(messages)
print(output)

```

```python
from langchain_core.output_parsers import JsonOutputParser
from langchain_core.pydantic_v1 import BaseModel, Field
from langchain_core.prompts import PromptTemplate
```

```python
class Joke(BaseModel):
    setup: str = Field(description="question to set up a joke")
    punchline: str = Field(description="answer to resolve the joke")

parser = JsonOutputParser(pydantic_object=Joke)
```

```python
prompt = PromptTemplate(
    template="Answer the user query.\n{format_instructions}\n{query}\n",
    input_variables=["query"],
    partial_variables={"format_instructions":parser.get_format_instructions()}
)
```

```python
chain = prompt | llm | parser
joke_query = "Tell me a joke"
output = chain.invoke({"query":joke_query})
# parser_output = parser.invoke(output)
print(output)
```

# Test chroma

```python
from langchain_chroma import Chroma
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.document_loaders import PyPDFLoader
```

```python
pdf_url = "https://arxiv.org/pdf/2307.09288.pdf"
pdf_loader = PyPDFLoader(pdf_url)
docs = pdf_loader.load()
chunk_size = 300
chunk_overlap = 0
splitter =  RecursiveCharacterTextSplitter(
    chunk_size=chunk_size,
    chunk_overlap=chunk_overlap,
    length_function = len,
    is_separator_regex=False
)
docs = splitter.split_documents(docs)
embedding_model = HuggingFaceEmbeddings()
chroma_db = Chroma.from_documents(
    documents=docs,
    embedding=embedding_model,
)
query = "what is multimodal in large language model"
similar_doc = chroma_db.similarity_search(query,k=4)
print(similar_doc[0])
```
